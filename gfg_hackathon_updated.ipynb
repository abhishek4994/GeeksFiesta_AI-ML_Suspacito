Removing all rows which contain outliers-

```python
indexNames = datanew[ (datanew['day'] < 0.0) ].index
datanew.drop(indexNames , inplace=True)
indexNames = datanew[ (datanew['hour'] < 0.0) ].index
datanew.drop(indexNames , inplace=True)
indexNames = datanew[ (datanew['PM2.5'] > 900.0) ].index
datanew.drop(indexNames , inplace=True)
indexNames = datanew[ (datanew['wind_speed'] < 0.0) ].index
datanew.drop(indexNames , inplace=True)
datanew
```
